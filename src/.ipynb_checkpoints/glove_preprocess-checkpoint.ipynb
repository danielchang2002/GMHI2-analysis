```python
import pandas as pd
import numpy as np
from utils import get_species, get_labels
```

```python
X = get_species()
X
```

```python
X_renamed = X.copy()
X_renamed.columns = list(map(lambda x : x.split("|")[-1], X.columns))
X_renamed
```

```python
from tqdm import tqdm
c = 0.00001

m, n = X_renamed.shape

with open("../data/corpus.txt", 'f') as f
    for i in tqdm(range(m)):
        bool_vec = X_renamed.iloc[i, :].values > c
        for col in X_renamed.columns[bool_vec]:
            f.write(col + " ")
        print("bruh " * n)
```
