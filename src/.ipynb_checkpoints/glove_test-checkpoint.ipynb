```python
import pandas as pd
import numpy as np
```

```python
word_arr = []
word_to_arr = {}

with open("../data/glove/vectors.txt") as f:
    lines = f.readlines()
    for line in lines:
        word = line.split()[0]
        word_arr.append(word)
        arr = [float(s) for s in line.split()[1:]]
        word_to_arr[word] = arr
```

```python
a = np.array(word_to_arr["s__Bacteroides_uniformis"])
b = np.array(word_to_arr["s__Escherichia_coli"])
np.linalg.norm(a - b)
```

```python
W = np.array([ele[1] for ele in word_arr[2:-1]])
W.shape
```

```python
W_norm = np.divide(W, np.linalg.norm(W, axis=1, keepdims=True))
```

```python
np.linalg.norm(W_norm[0, :])
```

```python
from utils import get_species
```

```python
X = get_species()
X
```

```python
X_renamed = X.copy()
X_renamed.columns = list(map(lambda x : x.split("|")[-1], X.columns))
X_renamed
```

```python
X_reduced = X_renamed[[ele[0] for ele in word_arr[2:-1]]]
X_reduced
```

```python
c = 0.00001
# num_words = np.sum((X_reduced > c).values, axis=1, keepdims=True)
X_trans = (X_reduced > c) @ W_norm
# X_trans = np.divide(X_trans, num_words)
X_trans
```

```python
X_trans.iloc[:, 4].plot.hist(bins=100)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s = StandardScaler()
X_trans_standard = s.fit_transform(X_trans)
X_trans_standard = np.nan_to_num(X_trans_standard)
X_trans_standard
```

```python
import matplotlib.pyplot as plt

plt.hist(X_trans_standard[:, 8], bins=100)
print()
```

```python
from utils import get_labels
y = get_labels()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_trans_standard, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import balanced_accuracy_score

logreg = LogisticRegression(random_state=42, penalty="l1", solver="liblinear", C=1)
logreg.fit(X_train, y_train)
y_hat = logreg.predict(X_test)
balanced_accuracy_score(y_test, y_hat)
```
