```python
import pandas as pd
import numpy as np
from utils import get_species, get_labels, get_labels_all
```

```python
X, y, y_all = get_species(), get_labels(), get_labels_all()
```

```python
X = X[list(filter(lambda x : "virus" not in x and "unclassified" not in x, X.columns))]
```

```python
weights = ((y_all == "T2D") | (y_all == "Obesity") | (y_all == "Overweight") | (y_all == "Underweight")).values
X_red = X.iloc[~weights, :]
```

```python
X_renamed = X_red.copy()
X_renamed.columns = list(map(lambda x : x.split("|")[-1], X.columns))
X_renamed
```

```python
from tqdm import tqdm
c = 0.00001

m, n = X_renamed.shape

with open("../data/corpus.txt", 'w') as f:
    for i in tqdm(range(m)):
        bool_vec = X_renamed.iloc[i, :].values > c
        for col in X_renamed.columns[bool_vec]:
            f.write(col + " ")
        f.write("bruh " * n)
```
