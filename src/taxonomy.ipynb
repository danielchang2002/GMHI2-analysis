```python
from utils import get_taxonomy

X = get_taxonomy()
```

```python
X
```

```python
features = list(X.columns)
```

```python
features
```

```python
from utils import get_species

c = 0.00001
s = get_species()
count = 1.0 * (s > c)
```

```python
from tqdm import tqdm

for feature in tqdm(features):
    filtered = list(filter(lambda x : feature in x, s.columns))
    sum = count[filtered].sum(axis=1)
    X[feature] = sum
```

```python
X.to_csv("../data/counts.csv")
```
